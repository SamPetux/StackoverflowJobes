```python
import csv
from math import ceil

LANGS = ('Java', 'JavaScript', 'CSS', 'PHP', 'Python', 'Ruby', 'SQL', 'C#', 'C++', 'C', 'Perl')
```

### 2011 год

```python
def get_exp(exp):
  match exp:
    case '<2':
      return 2
    case '41310':
      return 4
    case '41435':
      return 8
    case _:
      return 11
```

```python
def get_comp(comp):
  match comp:
    case '<$20,000':
      return 10000
    case '$20,000 - $40,000':
      return 30000
    case '$40,000 - $60,000':
      return 50000
    case '$60,000 - $80,000':
      return 70000
    case '$80,000 - $100,000':
      return 90000
    case '$100,000 - $120,000':
      return 110000
    case '$120,000 - $140,000':
      return 130000
    case _:
      return 150000
```

```python
professions = {}

with open('/content/2011 Stack Overflow Survey Results.csv', encoding='latin-1') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    n = 0
    for row in reader:
      if n > 1:
        ProgrammingExperience = get_exp(row[3]) if row[3] else ''
        Occupation = row[6] if row[6] != 'Other' else ''
        Country = row[0]
        Compensation = get_comp(row[45]) if row[45] and row[45] != 'Student / Unemployed' else ''
        Languages = row[30:43]
        while '' in Languages:
          Languages.remove('')
        if ProgrammingExperience and Occupation and Country and Compensation and Languages:
          if Occupation not in professions:
                professions[Occupation] = []
          professions[Occupation].append(
              {'country/region': Country,
               'experience (years)': ProgrammingExperience,
               'languages': Languages,
               'compensation': Compensation})
        continue
      n += 1
```

```python
with open('2011 summary file.csv', 'w', encoding='latin-1') as w_file:
  file_writer = csv.writer(w_file, delimiter = ',')
  file_writer.writerow(['profession', 'country/region', 'experience (years)',
                        'Java', 'JavaScript', 'CSS', 'PHP', 'Python', 'Ruby',
                        'SQL', 'C#', 'C++', 'C', 'Perl', 'Other',  'compensation'])
  n = 0
  for prof in professions:
    users = sorted(professions[prof], key=lambda d: d['country/region'])
    for user in users:
      us_languages = user['languages']
      if us_languages:
        languages = []
        for lang in LANGS:
          if lang in us_languages:
            languages.append('+')
            us_languages.remove(lang)
          else:
            languages.append('-')
        if us_languages:
          languages.append(', '.join(us_languages))
        else:
          languages.append('-')
        file_writer.writerow([prof, user['country/region'], user['experience (years)'], *languages, user['compensation']])
```

```python
def get_exp(exp):
  match exp:
    case 'Less than a year':
      return 1
    case '20 or more years':
      return 20
    case _:
      return int(exp.split()[2])
```

```python
def get_comp(comp):
  if comp < 5000:
    return 0
  if comp < 20000:
    return 10000
  if comp < 40000:
    return 30000
  if comp < 60000:
    return 50000
  if comp < 80000:
    return 70000
  if comp < 100000:
    return 90000
  if comp < 110000:
    return 110000
  if comp < 140000:
    return 130000
  if comp < 160000:
    return 150000
  return 170000
```

```python
def get_comp(total, frequent):
    if total and frequent:
      if frequent == 'Yearly':
        try:
          return ceil(float(total))
        except:
          return ''
      elif frequent == 'Yearly':
        try:
          return ceil(float(total)) * 12
        except:
          return ''
      else:
        try:
          return ceil(float(total)) * 52
        except:
          return ''
    return ''
```

```python
def processing(year: int) -> None:
  with open(f'{year} summary file.csv', encoding='latin-1') as fr, \
  open(f'{year} summary file processed.csv', 'w', encoding='latin-1') as fw:
    reader = csv.reader(fr, delimiter=',')
    writer = csv.writer(fw, delimiter=';')
    flag = False

    for row in reader:
      if flag:
        prof = row[0]
        if prof in PROFS and 5000 < float(row[-1]) < 150000:
          right_name = PROFS[prof]
          if isinstance(right_name, str):
            writer.writerow([right_name] + row[1:])
          else:
            for rn in right_name:
              writer.writerow([rn] + row[1:])
        continue

      writer.writerow(row)
      flag = True
```

```python
def adjust_for_inflation(salary: float,
                         year: int,
                         base_year: int=2024,
                         inflation_rates: dict=None) -> float:

    if inflation_rates is None:
      # Данные взяты с сайта
      # https://tradingeconomics.com/united-states/inflation-cpi
      inflation_rates = {
          2011: 0.0317,
          2012: 0.0208,
          2013: 0.0148,
          2014: 0.0163,
          2015: 0.0013,
          2016: 0.0127,
          2017: 0.0212,
          2018: 0.0245,
          2019: 0.0182,
          2020: 0.0123,
          2021: 0.0470,
          2022: 0.0802,
          2023: 0.0414,
          2024: 0.0296
          }

    if year == base_year:
      return salary

    adjusted_salary = salary

    for i in range(year, base_year):
      adjusted_salary *= (1 + inflation_rates[i + 1])

    return adjusted_salary
```

```python
def get_data() -> list:
  data = list()
  for year in range(2011, 2025):
    with open(f'{year} summary file processed.csv', encoding='latin-1') as f:
      reader = csv.reader(f, delimiter=';')
      flag = False
      for row in reader:
        if flag:
          drow = [year] + row + [adjust_for_inflation(float(row[-1]), year)]
          data.append(drow)
        flag = True
  return data
```
